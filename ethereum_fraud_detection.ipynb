```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "transaction_dataset.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "vagifa/ethereum-frauddetection-dataset",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:")
display(df.head())
```

```python
print("DataFrame Info:")
df.info()

print("\nDataFrame Description:")
display(df.describe())
```

```python
import numpy as np

# Drop unnecessary columns
df_processed = df.drop(columns=['Unnamed: 0', 'Index', 'Address'], errors='ignore')

# Identify columns with missing values (specifically ERC20 related ones)
missing_erc20_cols_numeric = [
    col for col in df_processed.columns if 'ERC20' in col and df_processed[col].dtype != 'object' and df_processed[col].isnull().any()
]
missing_erc20_cols_object = [
    col for col in df_processed.columns if 'ERC20' in col and df_processed[col].dtype == 'object' and df_processed[col].isnull().any()
]

# Impute missing numerical ERC20 values with 0
for col in missing_erc20_cols_numeric:
    df_processed[col] = df_processed[col].fillna(0)

# Impute missing categorical ERC20 values with 'Unknown'
for col in missing_erc20_cols_object:
    df_processed[col] = df_processed[col].fillna('Unknown')

# Verify that there are no more missing values in ERC20 columns
print("Missing values after imputation (ERC20 numeric):")
print(df_processed[missing_erc20_cols_numeric].isnull().sum())
print("\nMissing values after imputation (ERC20 categorical):")
print(df_processed[missing_erc20_cols_object].isnull().sum())

print("\nUpdated DataFrame Info after dropping columns and imputing missing values:")
df_processed.info()
```

```python
import pandas as pd

# Identify categorical columns for one-hot encoding
categorical_cols = df_processed.select_dtypes(include='object').columns

# Apply one-hot encoding
df_encoded = pd.get_dummies(df_processed, columns=categorical_cols, drop_first=True)

print("Shape of DataFrame after one-hot encoding:", df_encoded.shape)
print("\nFirst 5 records of encoded DataFrame:")
display(df_encoded.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of the target variable 'FLAG'
plt.figure(figsize=(6, 4))
sns.countplot(x='FLAG', data=df_encoded)
plt.title('Distribution of Fraud (FLAG) vs. Non-Fraudulent Transactions')
plt.xlabel('FLAG (0: Non-Fraudulent, 1: Fraudulent)')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df_encoded.drop('FLAG', axis=1)
y = df_encoded['FLAG']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nDistribution of target in training set:")
print(y_train.value_counts(normalize=True))
print("\nDistribution of target in test set:")
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize and train the Logistic Regression model
# Due to class imbalance, we set class_weight='balanced'
# Increased max_iter to address ConvergenceWarning
model = LogisticRegression(random_state=42, solver='liblinear', class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Model Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Visualize the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Fraud (0)', 'Fraud (1)'],
            yticklabels=['Non-Fraud (0)', 'Fraud (1)'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
